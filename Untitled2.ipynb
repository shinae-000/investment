```python
# @title 📈 실시간 시장 심리 분석 대시보드 (통합 버전) { display-mode: "form" }

# 1. 라이브러리 설치 (최초 1회만 실행하면 됩니다)
# !pip install -U finance-datareader pykrx

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import FinanceDataReader as fdr
import warnings
from sklearn.preprocessing import MinMaxScaler
from datetime import datetime, timedelta

# 환경 설정: 경고 무시 및 한글 폰트 설정
warnings.filterwarnings('ignore')
!apt-get -qq -y install fonts-nanum > /dev/null
fontpath = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
font_name = fm.FontProperties(fname=fontpath).get_name()
plt.rc('font', family=font_name)
plt.rcParams['axes.unicode_minus'] = False

# --- 인터페이스 설정 ---
시장_선택 = "KOSDAQ" # @param ["KOSPI", "KOSDAQ"]
분석_기간_개월 = 6 # @param {type:"slider", min:1, max:12, step:1}
업데이트_실행 = True # @param {type:"boolean"}

if 업데이트_실행:
    try:
        # 2. 데이터 자동 수집 (Naver/Investing.com 우회 경로)
        end_date = datetime.now().strftime("%Y-%m-%d")
        start_date = (datetime.now() - timedelta(days=500)).strftime("%Y-%m-%d")

        print(f"🔄 {시장_선택} 및 보조 지표 수집 중...")
        ticker = 'KOSPI' if 시장_선택 == "KOSPI" else 'KOSDAQ'
        df = fdr.DataReader(ticker, start_date, end_date).reset_index()
        df = df[['Date', 'Close']].rename(columns={'Close': 'Price'}).ffill()

        # 3. 피어앤그리드 지수 로직 계산
        # (1) 모멘텀: 125일 이격도
        df['MA125'] = df['Price'].rolling(window=125).mean()
        df['Momentum'] = (df['Price'] - df['MA125']) / df['MA125']

        # (2) RSI: 상대강도지수 (10일)
        delta = df['Price'].diff()
        gain = (delta.where(delta > 0, 0)).rolling(window=10).mean()
        loss = (-delta.where(delta < 0, 0)).rolling(window=10).mean()
        df['RSI_10'] = 100 - (100 / (1 + (gain / loss + 1e-9)))

        # (3) 스코어링 및 정규화 (0~100)
        scaler = MinMaxScaler(feature_range=(0, 100))
        df[['Mom_Score', 'RSI_Score']] = scaler.fit_transform(df[['Momentum', 'RSI_10']])
        df['FG_Index'] = (df['Mom_Score'] * 0.5 + df['RSI_Score'] * 0.5)

        # (4) 오실레이터 (추세 확인용)
        df['Short_EMA'] = df['FG_Index'].ewm(span=12, adjust=False).mean()
        df['Long_EMA'] = df['FG_Index'].ewm(span=26, adjust=False).mean()
        df['Oscillator'] = df['Short_EMA'] - df['Long_EMA']

        # 4. 그래프 시각화
        recent_df = df[df['Date'] >= (df['Date'].max() - pd.DateOffset(months=분석_기간_개월))]

        plt.style.use('seaborn-v0_8-whitegrid')
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10), sharex=True, gridspec_kw={'height_ratios': [1.2, 1]})

        # 상단: 가격 차트
        ax1.plot(recent_df['Date'], recent_df['Price'], color='#2c3e50', linewidth=2, label=f'{시장_선택} 종가')
        ax1.set_title(f"{시장_선택} 실시간 시장 심리 분석 리포트", fontsize=18, pad=20)
        ax1.legend(loc='upper left')
        ax1.grid(True, alpha=0.3)

        # 하단: 오실레이터 (공포/탐욕)
        ax2.fill_between(recent_df['Date'], recent_df['Oscillator'], 0,
                         where=(recent_df['Oscillator'] >= 0), color='#e74c3c', alpha=0.6, label='탐욕(Greed)')
        ax2.fill_between(recent_df['Date'], recent_df['Oscillator'], 0,
                         where=(recent_df['Oscillator'] < 0), color='#3498db', alpha=0.6, label='공포(Fear)')
        ax2.axhline(0, color='black', linewidth=1.2)
        ax2.set_ylabel('심리 추세 (Oscillator)')
        ax2.legend(loc='upper left')

        plt.tight_layout()
        plt.show()

        # 5. 전략 리포트 출력 (논리 보정 완료)
        curr = df.iloc[-1]
        prev = df.iloc[-2]
        score = float(curr['FG_Index'])
        c_osc = float(curr['Oscillator'])
        p_osc = float(prev['Oscillator'])

        # 상태 판정
        if score >= 75: state, color, mood = "극심한 탐욕", "🔴", "Greed"
        elif score >= 55: state, color, mood = "탐욕", "🟠", "Greed"
        elif score >= 45: state, color, mood = "중립", "🟡", "Neutral"
        elif score >= 25: state, color, mood = "공포", "🟢", "Fear"
        else: state, color, mood = "극심한 공포", "🔵", "Fear"

        print("\n" + "="*65)
        print(f"        📊 {시장_선택} 투자 심리 정밀 리포트")
        print("="*65)
        print(f" ✅ 분석 기준일 : {curr['Date'].strftime('%Y-%m-%d')}")
        print(f" ✅ 현재 점수   : {score:.2f} / 100")
        print(f" ✅ 현재 상태   : {color} {state}")
        print("-" * 65)
        print(" [실시간 투자 시그널]")

        if mood == "Greed":
            if c_osc > p_osc:
                print(" 🚀 매수 강세: 탐욕 구간이며 상승 에너지가 폭발 중입니다.")
            else:
                print(" ↗️ 보유 유효: 탐욕 구간이나 상승세가 둔화 중입니다. 익절 준비.")
        elif mood == "Fear":
            if c_osc > p_osc:
                print(" 📉 반등 조짐: 공포 구간이나 심리가 회복 중입니다. 저점 매수 기회.")
            else:
                print(" 🧊 관망 유지: 공포 구간이며 하락세가 지속 중입니다. 바닥 확인 필수.")
        else:
            print(" 🟡 방향성 탐색: 중립 구간입니다. 확실한 돌파 시점을 기다리세요.")

        if score <= 20:
            print("\n 📢 [역발상] 시장이 '패닉' 상태입니다. 공포에 살 때입니다.")
        elif score >= 80:
            print("\n 📢 [경고] 시장이 '광기' 상태입니다. 비중 축소를 권장합니다.")
        print("="*65)

    except Exception as e:
        print(f"❌ 오류 발생: {e}")
        print("💡 팁: '!pip install -U finance-datareader' 실행 후 세션을 다시 시작해 보세요.")
```

```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import ipywidgets as widgets
import json
from bs4 import BeautifulSoup

# 1. 종목명 또는 코드로 종목 정보를 가져오는 함수
def get_stock_info(search_term):
    # 만약 숫자로 된 6자리 코드라면 바로 반환
    if search_term.isdigit() and len(search_term) == 6:
        return search_term, f"Code:{search_term}"

    # 네이버 증권 종목 검색 API (한글 인코딩 보완)
    search_url = f"https://ac.finance.naver.com/ac?q={search_term}&q_enc=utf-8&st=111&frm=stock&r_format=json&r_enc=utf-8&r_unicode=1&t_koreng=1"

    try:
        headers = {'User-Agent': 'Mozilla/5.0'}
        response = requests.get(search_url, headers=headers)
        data = response.json()

        # 검색 결과 추출
        items = data['items'][0]
        if items:
            code = items[0][0][0] # 종목코드
            name = items[0][1][0] # 종목명
            return code, name
    except Exception as e:
        pass

    return None, None

# 2. 데이터 수집 및 시각화 함수
def analyze_stock(search_term, months):
    code, name = get_stock_info(search_term)

    if not code:
        print(f"'{search_term}'을(를) 찾을 수 없습니다. 종목코드 '445680'을 직접 입력해보세요.")
        return

    # 한 페이지당 10일치 데이터 (1개월당 2~3페이지 수집)
    pages = months * 3
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36'}
    df_list = []

    print(f"[{name} ({code})] 데이터를 불러오는 중... (약 {months}개월)")

    for page in range(1, pages + 1):
        url = f"https://finance.naver.com/item/frgn.naver?code={code}&page={page}"
        res = requests.get(url, headers=headers)
        tables = pd.read_html(res.text)

        if len(tables) > 2:
            temp_df = tables[2]
            if temp_df.dropna().empty: break # 데이터가 없으면 중단
            df_list.append(temp_df)
        else:
            break

    if not df_list:
        print("데이터 수집에 실패했습니다.")
        return

    df = pd.concat(df_list).dropna().reset_index(drop=True)
    df.columns = ['날짜', '종가', '전일비', '등락률', '거래량', '기관순매매량', '외국인순매매량', '외국인보유주수', '외국인보유율']

    # 데이터 타입 변환
    df['날짜'] = pd.to_datetime(df['날짜'])
    df['외국인보유율'] = df['외국인보유율'].astype(str).str.replace('%', '').astype(float)
    df = df.sort_values('날짜')

    # 3. 그래프 시각화 (상단: 주가, 하단: 수급)
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10), sharex=True, gridspec_kw={'height_ratios': [2, 1]})

    # [상단 서브플롯]
    ax1.plot(df['날짜'], df['종가'], color='black', linewidth=1.5, label='Price')
    ax1_twin = ax1.twinx()
    ax1_twin.plot(df['날짜'], df['외국인보유율'], color='red', linestyle='--', alpha=0.7, label='Foreign %')

    ax1.set_title(f"Analysis: {name} ({code})", fontsize=16, pad=20)
    ax1.set_ylabel('Stock Price (KRW)')
    ax1_twin.set_ylabel('Foreign Ownership (%)')
    ax1.grid(True, alpha=0.2)
    ax1.legend(loc='upper left')
    ax1_twin.legend(loc='upper right')

    # [하단 서브플롯]
    ax2.bar(df['날짜'], df['외국인순매매량'], color='royalblue', alpha=0.7, label='Foreign Net')
    ax2.bar(df['날짜'], df['기관순매매량'], color='orange', alpha=0.7, label='Inst. Net')
    ax2.axhline(0, color='black', linewidth=0.8)
    ax2.set_ylabel('Net Volume')
    ax2.grid(True, alpha=0.2)
    ax2.legend(loc='upper left')

    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# 4. 인터랙티브 실행 UI
stock_input = widgets.Text(value='445680', description='종목명/코드:')
month_slider = widgets.IntSlider(value=3, min=1, max=12, description='기간(개월):')
run_button = widgets.interact_manual(analyze_stock, search_term=stock_input, months=month_slider)
run_button.widget.children[2].description = "분석 시작" # 버튼 이름 변경
```

```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import ipywidgets as widgets
import json

def get_stock_info(search_term):
    if search_term.isdigit() and len(search_term) == 6:
        return search_term, f"Code:{search_term}"
    search_url = f"https://ac.finance.naver.com/ac?q={search_term}&q_enc=utf-8&st=111&frm=stock&r_format=json&r_enc=utf-8&r_unicode=1&t_koreng=1"
    try:
        headers = {'User-Agent': 'Mozilla/5.0'}
        response = requests.get(search_url, headers=headers)
        data = response.json()
        items = data['items'][0]
        if items:
            return items[0][0][0], items[0][1][0]
    except: pass
    return None, None

def analyze_stock(search_term, months):
    code, name = get_stock_info(search_term)
    if not code:
        print(f"'{search_term}'을(를) 찾을 수 없습니다.")
        return

    pages = months * 3
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36'}
    df_list = []

    for page in range(1, pages + 1):
        url = f"https://finance.naver.com/item/frgn.naver?code={code}&page={page}"
        res = requests.get(url, headers=headers)
        tables = pd.read_html(res.text)
        if len(tables) > 2:
            temp_df = tables[2].dropna()
            if temp_df.empty: break
            df_list.append(temp_df)
        else: break

    df = pd.concat(df_list).reset_index(drop=True)
    df.columns = ['날짜', '종가', '전일비', '등락률', '거래량', '기관순매매량', '외국인순매매량', '외국인보유주수', '외국인보유율']
    df['날짜'] = pd.to_datetime(df['날짜'])
    df['외국인보유율'] = df['외국인보유율'].astype(str).str.replace('%', '').astype(float)
    df = df.sort_values('날짜')

    # --- 누적 순매수량 계산 ---
    df['외국인누적'] = df['외국인순매매량'].cumsum()
    df['기관누적'] = df['기관순매매량'].cumsum()

    # 3단 그래프 구성 (Price / Daily Supply / Cumulative Supply)
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(14, 15), sharex=True,
                                        gridspec_kw={'height_ratios': [2, 1, 1.5]})

    # [1단] 주가 및 외국인 보유율
    ax1.plot(df['날짜'], df['종가'], color='black', linewidth=1.5, label='Price')
    ax1_t = ax1.twinx()
    ax1_t.plot(df['날짜'], df['외국인보유율'], color='red', linestyle='--', alpha=0.6, label='Foreign %')
    ax1.set_title(f"Analysis: {name} ({code})", fontsize=16, pad=20)
    ax1.set_ylabel('Stock Price')
    ax1.legend(loc='upper left'); ax1_t.legend(loc='upper right')
    ax1.grid(True, alpha=0.2)

    # [2단] 일별 수급 (막대)
    ax2.bar(df['날짜'], df['외국인순매매량'], color='royalblue', alpha=0.6, label='Foreign Daily')
    ax2.bar(df['날짜'], df['기관순매매량'], color='orange', alpha=0.6, label='Inst. Daily')
    ax2.axhline(0, color='black', linewidth=0.8)
    ax2.set_ylabel('Daily Net')
    ax2.legend(loc='upper left')
    ax2.grid(True, alpha=0.2)

    # [3단] 누적 수급 (선)
    ax3.plot(df['날짜'], df['외국인누적'], color='blue', linewidth=2, label='Foreign Cumul.')
    ax3.plot(df['날짜'], df['기관누적'], color='darkorange', linewidth=2, label='Inst. Cumul.')
    ax3.axhline(0, color='black', linewidth=0.8)
    ax3.set_ylabel('Cumulative Net')
    ax3.set_xlabel('Date')
    ax3.legend(loc='upper left')
    ax3.grid(True, alpha=0.2)

    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# 인터랙티브 실행
stock_input = widgets.Text(value='445680', description='종목명/코드:')
month_slider = widgets.IntSlider(value=3, min=1, max=12, description='기간(개월):')
ui = widgets.interact_manual(analyze_stock, search_term=stock_input, months=month_slider)
ui.widget.children[2].description = "분석 시작"
```
