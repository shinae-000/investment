```python
import os
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import pandas as pd
import requests
from IPython.display import display, HTML, clear_output
import warnings
import time

# --- ì‚¬ìš©ì ì„¤ì • (Form) ---
target_name = "373220" # @param {type:"string"}
months = 6 # @param {type:"slider", min:1, max:12, step:1}

def setup_korean_font():
    if not os.path.exists('/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'):
        os.system('apt-get -qq -y install fonts-num > /dev/null')
    fe = fm.FontEntry(fname=r'/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf', name='NanumBarunGothic')
    fm.fontManager.ttflist.insert(0, fe)
    plt.rc('font', family='NanumBarunGothic')
    plt.rcParams['axes.unicode_minus'] = False

def get_stock_info(search_term):
    search_term = search_term.strip()
    if search_term.isdigit() and len(search_term) == 6:
        return search_term, f"ì½”ë“œ:{search_term}"
    url = f"https://ac.finance.naver.com/ac?q={search_term}&q_enc=utf-8&st=111&frm=stock&r_format=json&r_enc=utf-8&r_unicode=1&t_koreng=1"
    try:
        response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
        items = response.json()['items'][0]
        if items: return items[0][0][0], items[0][1][0]
    except: pass
    return None, None

def calculate_rsi(df, period=14):
    delta = df['ì¢…ê°€'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))

def run_dashboard():
    setup_korean_font()
    warnings.filterwarnings('ignore')
    clear_output(wait=True)

    code, name = get_stock_info(target_name)
    if not code:
        display(HTML("<b style='color:red;'>ì¢…ëª©ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.</b>"))
        return

    try:
        # ë°ì´í„° ìˆ˜ì§‘ (ì§€í‘œ ê³„ì‚°ì„ ìœ„í•´ ì¶©ë¶„í•œ ë°ì´í„° í™•ë³´)
        pages = max(months * 4, 10)
        df_list = []
        headers = {'User-Agent': 'Mozilla/5.0'}

        for page in range(1, pages + 1):
            url = f"https://finance.naver.com/item/frgn.naver?code={code}&page={page}"
            res = requests.get(url, headers=headers)
            tables = pd.read_html(res.text)
            target_table = next((t for t in tables if t.shape[1] == 9 and 'ì¢…ê°€' in str(t.columns)), None)
            if target_table is not None:
                target_table.columns = ['ë‚ ì§œ', 'ì¢…ê°€', 'ì „ì¼ë¹„', 'ë“±ë½ë¥ ', 'ê±°ë˜ëŸ‰', 'ê¸°ê´€ìˆœë§¤ë§¤ëŸ‰', 'ì™¸êµ­ì¸ìˆœë§¤ë§¤ëŸ‰', 'ì™¸êµ­ì¸ë³´ìœ ì£¼ìˆ˜', 'ì™¸êµ­ì¸ë³´ìœ ìœ¨']
                target_table = target_table[target_table['ë‚ ì§œ'].str.contains(r'\d{4}\.\d{2}\.\d{2}', na=False)]
                df_list.append(target_table)
            else: break
            time.sleep(0.05)

        df = pd.concat(df_list).reset_index(drop=True)
        for col in ['ì¢…ê°€', 'ê±°ë˜ëŸ‰', 'ê¸°ê´€ìˆœë§¤ë§¤ëŸ‰', 'ì™¸êµ­ì¸ìˆœë§¤ë§¤ëŸ‰']:
            df[col] = pd.to_numeric(df[col], errors='coerce')
        df['ë‚ ì§œ'] = pd.to_datetime(df['ë‚ ì§œ'])
        df = df.sort_values('ë‚ ì§œ')

        # --- ì§€í‘œ ê³„ì‚° ---
        df['MA20'] = df['ì¢…ê°€'].rolling(20).mean()
        df['Upper'] = df['MA20'] + (df['ì¢…ê°€'].rolling(20).std() * 2)
        df['Lower'] = df['MA20'] - (df['ì¢…ê°€'].rolling(20).std() * 2)
        df['PV'] = df['ì¢…ê°€'] * df['ê±°ë˜ëŸ‰']
        df['VWMA60'] = df['PV'].rolling(window=60).sum() / df['ê±°ë˜ëŸ‰'].rolling(window=60).sum()
        df['RSI'] = calculate_rsi(df)
        df['ì™¸ì¸ëˆ„ì '] = df['ì™¸êµ­ì¸ìˆœë§¤ë§¤ëŸ‰'].cumsum()
        df['ê¸°ê´€ëˆ„ì '] = df['ê¸°ê´€ìˆœë§¤ë§¤ëŸ‰'].cumsum()

        # ì‹œê°í™” (3ë‹¨ êµ¬ì„±)
        plot_df = df.tail(months * 20)
        fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 12), sharex=True,
                                            gridspec_kw={'height_ratios': [2, 1, 0.8]})

        # 1. ê°€ê²© & VWMA60
        ax1.plot(plot_df['ë‚ ì§œ'], plot_df['ì¢…ê°€'], color='#333333', label='í˜„ì¬ê°€', alpha=0.8)
        ax1.plot(plot_df['ë‚ ì§œ'], plot_df['VWMA60'], color='#9467bd', label='60ì¼ VWMA(ê°€ì¤‘ì´í‰)', linewidth=2)
        ax1.fill_between(plot_df['ë‚ ì§œ'], plot_df['Lower'], plot_df['Upper'], color='gray', alpha=0.1)
        ax1.set_title(f"ğŸ“Š {name} ({code}) ì¢…í•© ë¶„ì„ ëŒ€ì‹œë³´ë“œ", fontsize=16)
        ax1.legend(loc='upper left')
        ax1.grid(True, alpha=0.3)

        # 2. ëˆ„ì  ìˆ˜ê¸‰
        ax2.plot(plot_df['ë‚ ì§œ'], plot_df['ì™¸ì¸ëˆ„ì '], label='ì™¸êµ­ì¸', color='#d62728', linewidth=2)
        ax2.plot(plot_df['ë‚ ì§œ'], plot_df['ê¸°ê´€ëˆ„ì '], label='ê¸°ê´€', color='#1f77b4', linewidth=2)
        ax2.axhline(0, color='black', linewidth=0.8)
        ax2.set_ylabel("ëˆ„ì  ìˆ˜ê¸‰ëŸ‰")
        ax2.legend(loc='upper left')
        ax2.grid(True, alpha=0.3)

        # 3. RSI
        ax3.plot(plot_df['ë‚ ì§œ'], plot_df['RSI'], color='#2ca02c', label='RSI(14)')
        ax3.axhline(70, color='#d62728', linestyle='--', alpha=0.5) # ê³¼ë§¤ìˆ˜ì„ 
        ax3.axhline(30, color='#1f77b4', linestyle='--', alpha=0.5) # ê³¼ë§¤ë„ì„ 
        ax3.fill_between(plot_df['ë‚ ì§œ'], 30, 70, color='#2ca02c', alpha=0.05)
        ax3.set_ylim(0, 100)
        ax3.set_ylabel("RSI")
        ax3.legend(loc='upper left')

        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

        # --- AI í•´ì„¤ ë¦¬í¬íŠ¸ ---
        last = df.iloc[-1]
        rsi_val = last['RSI']

        rsi_desc = "ğŸ“ˆ <b>ê³¼ë§¤ìˆ˜</b> (ë§¤ë„ ê³ ë ¤)" if rsi_val >= 70 else \
                   "ğŸ“‰ <b>ê³¼ë§¤ë„</b> (ë§¤ìˆ˜ ê²€í† )" if rsi_val <= 30 else "Neutral (ë³´í†µ)"

        supply_status = "ì™¸ì¸/ê¸°ê´€ ë™ë°˜ ë§¤ìˆ˜" if (last['ì™¸êµ­ì¸ìˆœë§¤ë§¤ëŸ‰'] > 0 and last['ê¸°ê´€ìˆœë§¤ë§¤ëŸ‰'] > 0) else \
                        "ì™¸ì¸ ë§¤ìˆ˜/ê¸°ê´€ ë§¤ë„" if (last['ì™¸êµ­ì¸ìˆœë§¤ë§¤ëŸ‰'] > 0 and last['ê¸°ê´€ìˆœë§¤ë§¤ëŸ‰'] < 0) else \
                        "ì™¸ì¸ ë§¤ë„/ê¸°ê´€ ë§¤ìˆ˜" if (last['ì™¸êµ­ì¸ìˆœë§¤ë§¤ëŸ‰'] < 0 and last['ê¸°ê´€ìˆœë§¤ë§¤ëŸ‰'] > 0) else "ìˆ˜ê¸‰ ì´íƒˆ ì£¼ì˜"

        display(HTML(f"""
        <div style="background-color:#f4f7f6; padding:20px; border-radius:12px; border:1px solid #dee2e6;">
            <h3 style="color:#2c3e50; margin-top:0;">ğŸ” ì¢…í•© ë¶„ì„ ë¦¬í¬íŠ¸</h3>
            <div style="display: flex; gap: 20px;">
                <div style="flex: 1; background: white; padding: 15px; border-radius: 8px;">
                    <h4 style="margin-top:0; color:#2ca02c;">1. RSI ì§€í‘œ</h4>
                    <p>í˜„ì¬ RSI: <b>{rsi_val: .2f}</b></p>
                    <p>ìƒíƒœ: {rsi_desc}</p>
                    <small>* 70 ì´ìƒì€ ê³¼ì—´, 30 ì´í•˜ëŠ” ë°”ë‹¥ ì‹ í˜¸ë¡œ í•´ì„í•©ë‹ˆë‹¤.</small>
                </div>
                <div style="flex: 1; background: white; padding: 15px; border-radius: 8px;">
                    <h4 style="margin-top:0; color:#d62728;">2. ìˆ˜ê¸‰ ë° ì¶”ì„¸</h4>
                    <p>ë‹¹ì¼ ìˆ˜ê¸‰: <b>{supply_status}</b></p>
                    <p>VWMA ì¶”ì„¸: <b>{"ê°•ì„¸ ìœ ì§€" if last['ì¢…ê°€'] > last['VWMA60'] else "ì¡°ì • êµ­ë©´"}</b></p>
                    <small>* ê°€ê²©ì´ VWMA60 ìœ„ì— ìˆìœ¼ë©´ ì§€ì§€ì„¸ê°€ ê°•í•¨ì„ ì˜ë¯¸í•©ë‹ˆë‹¤.</small>
                </div>
            </div>
        </div>
        """))

    except Exception as e:
        print(f"ì˜¤ë¥˜ ë°œìƒ: {e}")

run_dashboard()
```
